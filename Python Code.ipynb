```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.offsetbox import OffsetImage, AnnotationBbox

# Read sensor data from CSV file
locations_data = pd.read_csv("GrowLocations.csv")

# Cleaning Data
# Swap the column names 'Latitude' and 'Longitude'
locations_data.rename(columns={'Latitude': 'Temp_Latitude', 'Longitude': 'Latitude'}, inplace=True)
locations_data.rename(columns={'Temp_Latitude': 'Longitude'}, inplace=True)

# Filter data based on latitude and longitude boundaries
filtered_locations = locations_data[
    (locations_data['Latitude'] >= 50.681) & 
    (locations_data['Latitude'] <= 57.985) & 
    (locations_data['Longitude'] >= -10.592) & 
    (locations_data['Longitude'] <= 1.6848)
]

# Map Plotting
# Load the UK map
uk_map = plt.imread("map7.png")

# Create Matplotlib figure and axis
fig, ax = plt.subplots(figsize=(10, 10))

# Display the UK map with given boundaries
ax.imshow(uk_map, extent=[-10.592, 1.6848, 50.681, 57.985])

# Plot sensor locations on the map
for index, row in filtered_locations.iterrows():
    x, y = row['Longitude'], row['Latitude']
    ax.plot(x, y, marker='o', markersize=5, color='purple')

plt.show()
```
